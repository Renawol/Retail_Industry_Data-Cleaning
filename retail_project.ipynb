*A partir de la columna 16 vienen otra vez todas las horas por mes pero acumuladas

Atendiendo a lo anterior, este archivo solo contiene las horas totales mes por mes de cada tienda como información adicional. Por tanto, lo que nos interesa es añadir esto a nuestra tabla de sales para tener ahí toda la información recogida.

```python
openings.columns = openings.iloc[4]
```

```python
openings = openings.iloc[5:, :]
```

```python
openings.reset_index(drop= True, inplace= True)
```

```python
openings.head()
```

```python
sales['monthyear'] = sales['month'] + ' ' + sales['year'].astype(str)
```

```python
sales['monthyear'] = pd.to_datetime(sales['monthyear'], format= '%B %Y')
```

```python
sales['id_monthyear'] = sales['storeid'].astype(str) + '_' + sales['monthyear'].dt.strftime('%#m.%Y')
```

Este será el identificador

```python
openings.columns
```

Vamos a quitar las columnas de horas acumuladas ya que no las necesitamos. Si fuera necesario se podría calcular de nuevo más tarde o volver a llamar.

De la misma manera vamos a quitar las columnas relativas a Julio, Agosto y Septiembre de 2017 porque en el archivo de sales no hemos recibido la información de esos meses.

```python
openings_fixed = openings.iloc[:, :13]
```

Ahora que ya tenemos la columna monthyear para usarla como identificador, vamos a derretir openings para establecer los meses y año como filas

```python
openings_melted = openings_fixed.melt(id_vars=['id', 'Store name', 'Region', 'Scheme'], var_name= 'monthyear', value_name= 'hoursopened_pm')
```

```python
openings_melted.head()
```

```python
openings_melted['id_monthyear'] = openings_melted['id'].astype(str) + '_' + openings_melted['monthyear'].astype(str)
```

```python
openings_melted.head()
```

```python
df_final = sales.merge(openings_melted[['id_monthyear', 'hoursopened_pm']], on= 'id_monthyear')
```

```python
df_final.head()
```

```python
depts = df_final[['dept_id', 'dept_name']]
```
